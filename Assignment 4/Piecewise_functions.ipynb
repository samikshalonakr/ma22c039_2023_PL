```python
#Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the number of points and the x values
num_points = 1000001
x_values = np.linspace(-5, 5, num_points)
```

```python
# Compute the function values
def f(x):
    return np.where(x < 0, x * np.sin(x**2), np.cos(x))

y_values = f(x_values)
```

```python
# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values)
plt.title('f(x) = x * sin(x^2) if x < 0, f(x) = cos(x) if x >= 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
