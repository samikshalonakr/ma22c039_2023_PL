```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Task 1: Graphing sin(x) for x in the interval [0, 10]
x = np.linspace(0, 10, 1000)  # Generate 1000 points between 0 and 10
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Graph of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 500)  # Generate 500 x values between -5 and 5
y = np.linspace(-5, 5, 500)  # Generate 500 y values between -5 and 5
x, y = np.meshgrid(x, y)  # Create a grid of (x, y) values

z = np.cos(x**2 + y**2)

# Create a 3D surface plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

plt.title('Graph of cos(x^2 + y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
